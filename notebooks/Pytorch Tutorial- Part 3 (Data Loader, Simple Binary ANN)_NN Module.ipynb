**DATASET & Dataloader**

**What is a Dataset?**
A Dataset is an abstract class representing a dataset. It provides a way to access your data samples one at a time.

You typically subclass torch.utils.data.Dataset and override:

__len__() – returns the number of samples

__getitem__(idx) – loads and returns a sample at index idx

**What is a DataLoader?**
A DataLoader wraps a Dataset and provides:

a. Batching

b. Shuffling

c. Parallel loading with multiprocessing

This is useful for efficient model training.

*** Why not just use tensors directly?***

Key Advantages:

| **Feature** | **Using Tensors Directly** | **Using Dataset & DataLoader** |
|-------------|-----------------------------|----------------------------------|
| Efficient batching | ❌ Manual and error-prone | ✅ Automatic |
| Shuffling data | ❌ Must be coded by hand | ✅ Built-in |
| Custom & Easy applying transformations (like image augmentation) | ❌ Difficult | ✅ Easy via `__getitem__` |
| Handling large datasets | ❌ Must fit into memory | ✅ Can load on-the-fly |
| Multi-process loading | ❌ No support | ✅ Via `num_workers` in DataLoader |

**Applying Transformations during the data fetching**

```python
from torchvision import transforms

class CustomDataset(Dataset):
  def __init__(self, X, y, transform=None):
    self.features = X
    self.labels = y
    self.transform=transform

  def __len__(self):
    return self.features.shape[0]

  def __getitem__(self,index):
    x = torch.tensor(self.features[index])
    y = torch.tensor(self.labels[index])

    x=x.float()
    y=y.float()
    if self.transform:
      x = self.transform(x)  # Apply transformation on feature only
    return x, y

# Define a built-in transformation (normalize each feature vector)
# Scale features between 0 and 1 using Lambda
#This is an anonymous function (lambda) that takes an input tensor x and normalizes it by dividing all elements by its maximum value
transform = transforms.Lambda(lambda x: x / torch.max(x))


dataset=CustomDataset(X_tensor,y_tensor,transform=transform)
dataset.__getitem__(1)
```

```python
dataloader = DataLoader(dataset, batch_size=10, shuffle=True)

# Step 5: Iterate through the DataLoader
for batch_x, batch_y in dataloader:
    print("Batch features:", batch_x)
    print("Batch labels:", batch_y)
```

**Simple ANN with Mini Batch Gradient Descent using Dataset and Dataloader Class**

```python
X_train_tensor, X_test_tensor, y_train_tensor, y_test_tensor = train_test_split(X_tensor, y_tensor, test_size=0.2, random_state=42)
```

```python
train_dataset = CustomDataset(X_train_tensor, y_train_tensor, transform=None)
test_dataset = CustomDataset(X_test_tensor, y_test_tensor, transform=None)
```

```python
train_dataset.__getitem__(10)
```

```python
train_loader=DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader=DataLoader(test_dataset,batch_size=32, shuffle=False)
```

```python
# create model class
import torch
import torch.nn as nn

class SimpleNN(nn.Module):
  def __init__(self, feature_size):
    super().__init__()
    self.input=nn.Linear(feature_size, 128)
    self.hidden_layer1=nn.Linear(128,64)
    self.hidden_layer1=nn.Linear(128,10)
    self.output_layer=nn.Linear(5,1)
    self.sigmoid=nn.Sigmoid()
    self.relu=nn.ReLU()


  def forward(self, x):
    x=self.input(x)
    x=self.relu(x)
    x=self.hidden_layer1(x)
    x=self.relu(x)
    out=self.output_layer(x)
    out=self.sigmoid(out)

    return out
```

```python
lr=0.0001
feature_size=X_train_tensor.shape[1]
epochs=300
loss_function=nn.BCELoss()
```

```python
model=SimpleNN(feature_size)
optimizer=torch.optim.SGD(model.parameters(), lr=lr)

#Now loop through Dataloader

for epoch in range(epochs):
  for batch_features, batch_labels in train_loader:

    #forward pass operation
    y_pred = model(batch_features) # Convert batch_features to float32
    #clear any gradient
    optimizer.zero_grad()

    #Evaluate Loss
    loss=loss_function(y_pred,batch_labels.view(-1,1))

    #backpropagation
    loss.backward()

    #Updating Parameters(weights and biases)

    optimizer.step()

  print(f"Epoch: {epoch} Loss: {loss.item()}")
```

```python
accuracy=0
with torch.no_grad():
  for test_batch_features, test_batch_labels in test_loader:
      y_pred=model.forward(test_batch_features)
      for i in range(len(y_pred)):
        if (y_pred[i]>0.5):
          y_pred[i]=1
        else:
          y_pred[i]=0

        if(y_pred[i]==y_test_tensor[i]):
          accuracy=accuracy+1

print("The overall accuracy is",(accuracy/y_test_tensor.shape[0])*100)

```
